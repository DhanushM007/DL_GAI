```python
# ================================
# IMPORTS
# ================================
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Input
import numpy as np
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt

# ================================
# LOAD DATA
# ================================
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)

# ================================
# DISPLAY SAMPLE IMAGES
# ================================
fig, axs = plt.subplots(2, 5, figsize=(12,6))

n = 0
for i in range(2):
    for j in range(5):
        axs[i, j].matshow(X_train[n])
        axs[i, j].set_title(y_train[n])
        axs[i, j].axis('off')
        n += 1

plt.show()

# ================================
# PREPROCESSING
# ================================
X_train = X_train.reshape(60000, 784).astype("float32") / 255
X_test = X_test.reshape(10000, 784).astype("float32") / 255

# ================================
# MODEL FUNCTION (Reusable)
# ================================
def build_model():
    model = Sequential(name="DFF-Model")
    model.add(Input(shape=(784,)))
    model.add(Dense(128, activation='relu', kernel_initializer='he_normal'))
    model.add(Dense(64, activation='relu', kernel_initializer='he_normal'))
    model.add(Dense(32, activation='relu', kernel_initializer='he_normal'))
    model.add(Dense(10, activation='softmax'))

    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

# ================================
# TRAIN & RECORD ACCURACY
# ================================
epoch_list = [5, 50, 100]
results = {}

for ep in epoch_list:
    print(f"\n======= Training for {ep} epochs =======")

    model = build_model()

    history = model.fit(
        X_train, y_train,
        epochs=ep,
        batch_size=32,
        validation_split=0.2,
        verbose=1
    )

    # Evaluate
    train_loss, train_acc = model.evaluate(X_train, y_train, verbose=0)
    test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)

    results[ep] = (train_acc, test_acc)

    print(f"\nEpochs: {ep}")
    print(f"Training Accuracy: {train_acc:.4f}")
    print(f"Test Accuracy: {test_acc:.4f}")

# ================================
# FINAL COMPARISON
# ================================
print("\n========= FINAL RESULTS =========")
for ep in results:
    print(f"Epochs: {ep} | Train Acc: {results[ep][0]:.4f} | Test Acc: {results[ep][1]:.4f}")
```

```python
# ==========================================
# IMPORTS
# ==========================================
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Input, Conv2D, MaxPooling2D, Flatten
import numpy as np
import matplotlib.pyplot as plt

# ==========================================
# MODEL DEFINITIONS
# ==========================================

# Dense Model (MNIST & Fashion MNIST)
def build_dense(input_size):
    model = Sequential()
    model.add(Input(shape=(input_size,)))
    model.add(Dense(128, activation='relu', kernel_initializer='he_normal'))
    model.add(Dense(64, activation='relu', kernel_initializer='he_normal'))
    model.add(Dense(32, activation='relu', kernel_initializer='he_normal'))
    model.add(Dense(10, activation='softmax'))

    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

# CNN Model (CIFAR-10)
def build_cnn():
    model = Sequential()
    model.add(Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)))
    model.add(MaxPooling2D((2,2)))

    model.add(Conv2D(64, (3,3), activation='relu'))
    model.add(MaxPooling2D((2,2)))

    model.add(Flatten())
    model.add(Dense(64, activation='relu'))
    model.add(Dense(10, activation='softmax'))

    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

# ==========================================
# DATASETS
# ==========================================

datasets = {
    "MNIST": keras.datasets.mnist.load_data(),
    "Fashion MNIST": keras.datasets.fashion_mnist.load_data(),
    "CIFAR-10": keras.datasets.cifar10.load_data()
}

epoch_list = [5, 50, 100]

# Store results в†’ {dataset: [(train_acc, test_acc), ...]}
results = {name: [] for name in datasets.keys()}

# ==========================================
# TRAINING LOOP
# ==========================================

for name, data in datasets.items():
    print(f"\n================ {name} =================")

    (X_train, y_train), (X_test, y_test) = data

    # --------------------------------------
    # PREPROCESSING
    # --------------------------------------
    if name == "CIFAR-10":
        y_train = y_train.flatten()
        y_test = y_test.flatten()

        X_train = X_train.astype("float32") / 255
        X_test = X_test.astype("float32") / 255
    else:
        X_train = X_train.reshape(X_train.shape[0], -1).astype("float32") / 255
        X_test = X_test.reshape(X_test.shape[0], -1).astype("float32") / 255

    # --------------------------------------
    # TRAIN FOR DIFFERENT EPOCHS
    # --------------------------------------
    for ep in epoch_list:
        print(f"\nTraining for {ep} epochs...")

        if name == "CIFAR-10":
            model = build_cnn()
        else:
            model = build_dense(X_train.shape[1])

        history = model.fit(
            X_train, y_train,
            epochs=ep,
            batch_size=32,
            validation_split=0.2,
            verbose=0
        )

        # --------------------------------------
        # EVALUATION
        # --------------------------------------
        train_acc = model.evaluate(X_train, y_train, verbose=0)[1]
        test_acc = model.evaluate(X_test, y_test, verbose=0)[1]

        results[name].append((train_acc, test_acc))

        print(f"Train Accuracy: {train_acc:.4f}")
        print(f"Test Accuracy : {test_acc:.4f}")

# ==========================================
# PRINT FINAL RESULTS TABLE
# ==========================================

print("\n\n=========== FINAL RESULTS TABLE ===========")

for name in results:
    print(f"\n{name}")
    print("Epochs | Train Acc | Test Acc")
    print("--------------------------------")
    for i, ep in enumerate(epoch_list):
        train_acc, test_acc = results[name][i]
        print(f"{ep:6} | {train_acc:.4f}   | {test_acc:.4f}")

# ==========================================
# PLOTTING COMPARATIVE GRAPH
# ==========================================

plt.figure()

for name in results:
    test_values = [x[1] for x in results[name]]
    plt.plot(epoch_list, test_values, marker='o', label=name)

plt.title("Comparative Analysis of MNIST, Fashion MNIST, CIFAR-10")
plt.xlabel("Epochs")
plt.ylabel("Test Accuracy")
plt.legend()
plt.grid()

plt.show()
```
